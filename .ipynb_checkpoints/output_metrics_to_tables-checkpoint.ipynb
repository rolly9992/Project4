```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_generic = pd.read_excel('generic_metrics.xlsx')
df_generic.head() 
```

```python
generic_stats = df_generic[['model','Accuracy','Precision','Recall','F1 Score']]
generic_stats=generic_stats.groupby('model').mean()
generic_stats
```

```python
generic_stats.mean()
```

```python
df_features_reduction = pd.read_excel('features_testing_metrics.xlsx')
df_features_reduction.head()
```

```python
num_features = df_features_reduction[['model','Accuracy','Precision','Recall','F1 Score','numberoffeatures']]
num_features_grouped = num_features.groupby(['model','numberoffeatures']).mean()
num_features_grouped
```

```python
avg_by_num_features = num_features[['Accuracy','Precision','Recall','F1 Score','numberoffeatures']].groupby(['numberoffeatures']).mean()
avg_by_num_features
```

```python
best_feature_num = df_features_reduction[['model','Accuracy','Precision','Recall','F1 Score','numberoffeatures']]
best_feature_num = best_feature_num[best_feature_num['numberoffeatures']==100]
avg_best_features = best_feature_num[['model','Accuracy','Precision','Recall','F1 Score']]
avg_best_features_grouped = avg_best_features.groupby(['model']).mean()
avg_best_features_grouped
```

```python
avg_best_features_grouped.mean()
```

```python
df_grid = pd.read_excel('all_grid_search_metrics.xlsx')
df_grid.head()
```

```python
grid_stats = df_grid[['model name','Accuracy','Precision','Recall','F1 Score']]
grid_stats_grouped=grid_stats.groupby('model name').mean()
grid_stats_grouped
```

```python
grid_stats_grouped.mean()
```

```python
s1 = pd.DataFrame(generic_stats.mean())
s1.columns = ['Generic']
s1
s2 = pd.DataFrame(avg_best_features_grouped.mean())
s2.columns = ['Reduced Features']
s2
s3 = pd.DataFrame(grid_stats_grouped.mean())
s3.columns=['Grid Search']
s3
sidebyside = pd.concat([s1,s2,s3],axis=1)
sidebyside
```
